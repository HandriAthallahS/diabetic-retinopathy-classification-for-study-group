```
from ultralytics import YOLO
from sklearn.metrics import classification_report
import os
from pathlib import Path

model = YOLO("runs/classify/ddr-224/weights/best.pt")

test_path = Path(dataset.location) / "test"

y_true = []
y_pred = []

for class_id, class_name in enumerate(sorted(os.listdir(test_path))):
    class_folder = test_path / class_name

    for img_path in class_folder.glob("*.*"):
        result = model(str(img_path), verbose=False)[0]

        pred_class = result.probs.top1

        y_pred.append(pred_class)
        y_true.append(class_id)

print(classification_report(y_true, y_pred, digits=4))
```
